```python
import sqlite3 # sqlite is a built-in package in python...!
import pandas as pd
import numpy as np
import os
```

```python
try:
    os.remove('demo.db')
except:
    pass
```

```python
# create a database called demo
db = sqlite3.connect('demo.db')
```

```python
asset = pd.read_csv('https://mfin6201.s3.amazonaws.com/company_assets.csv')
```

```python
asset.to_sql('asset',db,if_exists='replace',index=False)
```

```python
# Now let's query the database using SQL
pd.read_sql('SELECT permno,fyear,at FROM asset',db)
```

```python
pd.read_sql('SELECT * FROM asset limit 10',db)
```

```python
pd.read_sql('SELECT permno,fyear FROM asset WHERE at>10000',db)
```

```python
# get information for the database
pd.read_sql('''
    SELECT * FROM sqlite_master
''',db)
```

```python
left = pd.DataFrame({'A': [1,2,3,4],
                        'B': [11,22,33,44]})

right = pd.DataFrame({'A': [1,2,4,5],
                        'C': [11,22,44,55]})
```

```python
left.to_sql('left_table',db)
right.to_sql('right_table',db)
```

```python
pd.read_sql('SELECT * FROM left_table',db)
```

```python
pd.read_sql('''
    SELECT * FROM left_table l
    LEFT JOIN right_table r
    ON l.A = r.A
''',db)
```

```python
pd.read_sql('''
    SELECT l.A, B, C FROM left_table l
    JOIN right_table r
    ON l.A = r.A
''',db)
```
